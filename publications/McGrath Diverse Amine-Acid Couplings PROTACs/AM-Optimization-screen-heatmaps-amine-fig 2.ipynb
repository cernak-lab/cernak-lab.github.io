```python
x = ['Kartedts Catalyst \n DMSO','Kartedts Catalyst \n Bu$_2$O','PtCl$_2$ \n DMSO','PtCl$_2$ \n Bu$_2$O','BCF \n DMSO','BCF \n Bu$_2$O']
y = ['dPPP','tBu BrettPhos','BrettPhos', 'tBu XPhos']

# matplotlib notebook

sorted([f for f in matplotlib.font_manager.fontManager.ttflist], key=lambda f:f.name);
```

```python
#matplotlib.rcParams['mathtext.fontset']='custom'
matplotlib.rcParams['mathtext.rm']
vmax = 25
fig, ax = plt.subplots(figsize=(14,7))
im = ax.imshow(data,vmax=vmax)

# We want to show all ticks...
ax.set_xticks(np.arange(len(x)))
ax.set_yticks(np.arange(len(y)))
# ... and label them with the respective list entries
ax.set_xticklabels(x,fontname='Arial')
ax.set_yticklabels(y,fontname='Arial')


for i in range(len(x)):
    for j in range(len(y)):
        text = ax.text(i, j, format(data[j][i]),
                       ha="center", va="center", color="w", size=18, weight = "bold")

plt.axis(ymin=3.5,ymax=-0.5)

divider = make_axes_locatable(ax)
cax = divider.append_axes("right", size="5%", pad=0.05)
plt.colorbar(im, cax=cax, orientation = 'horizontal')

# plt.savefig('Larger text AM-amine opt assay yield scaled.png',dpi=600)
```
